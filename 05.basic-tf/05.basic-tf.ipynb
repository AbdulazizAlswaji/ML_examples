```python
def c_to_f(c):
    return c * 1.8 + 32
```

```python
c_to_f(100)
```

# ===========================================

```python
Cs = [0, 1, 4, 12, 30, 45, 100, 105, 90, 77,-40, -10,   8, 15, 22,  38]
Fs = []
for c in Cs:
    f = c_to_f(c)
    Fs.append(f)
```

```python
Fs
```

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(0.1))
```

c = 1 ==> f = 33.8 | p = 44.8 <br>
(33.8 - 44.8) ^ 2 = 121<br>

F = c * 2.8 + 42

---------------------------------

c = 10 ==> f = 50 | p = 50 <br>
(50 - 50) ^ 2 = 0 <br>

F = c * 1.8 + 32


```python
history = model.fit(Cs, Fs, epochs=500)
```

```python
c = 101.65
print('model prediction:',model.predict([c]))
print('actual value:',c_to_f(c))
```
